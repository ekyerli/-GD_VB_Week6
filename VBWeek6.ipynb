## 1. Regresyon vs Sınıflandırma
Regresyonun sınıflandırmadan nasıl farklı olduğu üzerinde biraz daha duralım.

Bir sınıflandırma problemi için hedef değişken kategorik'tir. Bu, değişkenin yalnızca belirli bir kümeden ayrık değerler alması anlamına gelir. Bu kümenin basit sürümleri {evet, hayır} veya {pozitif, negatif} gibi bir görünüme sahip olabilir. Sonuçlar, aynı zamanda, {yüksek, orta, düşük} veya {satın alma, kiralama, satılık değil} gibi ikiden fazla değer içerebilir. Kesin olarak sayılabilecek ve etiketlenebilecek hemen hemen her şey kategorik bir değişken olarak düşünülebilir.

Bunun aksine, regresyon problemleri sürekli sonuç değişkenine sahiptir. Örneğin, Apple’ın Nasdaq’daki hisselerinin ertesi gün kapanış fiyatını tahmin eden bir model oluşturmak isteyebilirsiniz. Açıkçası, Apple’ın hisse senedi fiyatı kategorik bir değişken değildir. Her ne kadar yukarıdan ve aşağıdan bir şekilde sınırlı olsa da, hala sonsuz sayıda olası değere sahiptir.

## Bir Model Tasarlamak:
İlk adım, hedef değişkenle bir ilişkisi olduğuna inandığımız özelliklerin başlangıç kümesini seçmektir. Futbol örneği sadece bir özellik içerir. Ancak, daha gerçekçi bir veri kümesinde, modelinize katkıda bulunabilecek birden fazla özelliğe sahip olacaksınız. İlk özellik kümesinin seçilmesi, tipik olarak model seçiminden önce yaptığınız özellik mühendisliği ve keşif analizinden kaynaklanmaktadır.

İki özelliğimiz ( x  ve  z ) ve bir hedefimiz ( y ) olduğunu varsayarsak, ilk modelimiz aşağıdaki şekilde ifade edilebilir:

y=α+βx+θz
 
Yukarıdaki formülde,  α  bias term,  β  ve  θ  ise katsayıdır.

## Katsayıların Tahmin Edilmesi
Bir sonraki adım, modelimizin en uygun parametrelerini bulmaktır. Bu, futbol örneğinde mavi çizgi çizmek ile aynı şeydir. Yukarıdaki tabloya daha yakından bakarsanız, mavi 'den başka bazı çizgiler olduğunu görürsünüz. Aslında, o grafiğe mavi çizgi 'ye az çok benzeyen sonsuz sayıda çizgi çizebiliriz.

## Doğrusal Regresyon Varsayımları:
Modelimizin parametrelerini tahmin ettikten sonra, verilerimiz ve modelimizle ilgili bazı varsayımların geçerli olduğundan emin olmanız gerekir. Özellikle, doğrusal regresyon kullanmak için modelimiz tarafından karşılanması gereken az sayıda varsayım vardır. Doğrusal regresyon altı koşulu varsayar:

    Özellik(ler) ve hedef(ler) arasındaki ilişki doğrusaldır.
    Modelin hataları ortalamada sıfıra eşit olmalıdır.
    Modelin hataları, tutarlı bir şekilde dağıtıktır ve heteroscedasticity olarak bilinir.
    Özellikler en fazla sadece zayıf bir şekilde koreledir. Farklı bir şekilde ifade edersek, güçlü bir multicollinearity yoktur.
    Modelin hataları birbiriyle korele olmamalıdır.
    Özellikler ve model hataları birbirinden bağımsızdır.
Bu varsayımlar muhtemelen şu anda size soyut geliyor, ama endişelenmeyin. Daha sonra bu bölümde, bu varsayımları gözden geçireceğiz ve modelimiz bu koşulları karşılamadığında bile doğrusal regresyon kullanmak için alabileceğimiz bazı yaklaşımları açıklayacağız. Sİz de ileride oluşturacağınız modellerde bu koşulları kontrol edip, uymayan koşullar için modelinizi düzeltmeye çalıştıktan sonra, bir sonraki adıma geçebilirsiniz.

## Özellikler ve hedef arasındaki ilişkiyi anlama:
Önceki adımlardan sonra, özellikler ile hedef değişken arasındaki ilişkiyi yakalayan özelliklerin katsayıları için bazı değerler elde ederiz. Hedef değişken ile her bir özellik arasındaki ilişkiyi tahmini katsayıları kullanarak nicelleştirebiliriz. Özellikle, mavi çizginin temsil ettiği tahmini fonksiyon şöyledir:

ortalamagolsayısı=−0.01+0.09Xşutsayısı
 
Bu, şut sayısı 10 artarsa, o zaman ortalama gol sayısının 0,9 arttığını gösterir.

##  Uygunluğun değerlendirilmesi:
Bir modelin sonuç değişkenindeki bilgiyi ne kadar iyi açıkladığını ölçmek için, modelin performans verileri üzerindeki performansına bakarız. Eğitim verileri, modeli tahmin etmek için kullandığımız kayıtlardan oluşmaktadır. Daha sonra bootcamp'in devamında göreceğimiz gibi sınıflandırma modellerinin performansını ölçmek için, doğruluk, kesinlik ve geri çağırma gibi metrikler oluşturmak için bir karışıklık matrisi (confusion matrix) kullanıyoruz.

Ancak, bir karışıklık matrisi oluşturmak, regresyon problemlerinde karşılaştığımız sürekli değişkenler için işe yaramaz. Sonsuz bir karışıklık matrisi sonsuz kafa karıştırıcıdır. Bunun yerine, modelimizin uygunluğunu değerlendirmek için R-kare veya ayarlanmış R-kare gibi metrikler kullanıyoruz (bu bölümün ilerleyen derslerinde ele alınacak). Modelimizin performansından memnun kalırsak, sonraki adımlara geçebiliriz. Ancak, modelimizin hedef değişkeni çok iyi açıklayamadığını düşünürsek, modelimizde bazı ekstra özellikler eklememiz ve 2. adıma (tahmin aşamasına) dönmemiz gerekir.

## Bilinmeyeni tahmin etmek:
Regresyon modelleri bilinmeyen hedef değerleri tahmin etmede oldukça esnektir. Bu, özellikler ve hedef değişken arasındaki ilişkiyi nasıl tahmin ettiğimizin bir sonucudur. Sonlu bir dizi gözlemlenmiş olguya dayanarak, prensipte herhangi bir gözlemlenmemiş değer hakkında bir tahmin yapabilen bir çizgi oluştururuz. Tahmin ile, daha önce gözlemlenmemiş gözlemler için sonuç değerlerini ortaya koyarız.

Test kümesinde, doğru hedef değer ile modelimizin ne olacağını tahmin ettiği arasındaki farka bakabiliriz. Tahminimizi bu şekilde düşündüğümüzde, modelimizin daha önce görülmemiş gözlemlerde ne kadar iyi performans göstereceğini objektif olarak belirlemenin bir yolunu sunar.



## Ölçme testi performansı:
Şut sayısına göre maç başına kaç gol atılabileceğimizi tahmin eden bir fonksiyonumuz var. Ancak bu tahmin uygulamada güvenilir olacak mı? Stratejimiz bu bilgilere dayanmalı mı? Bunu belirlemek için modelimizi bir test kümesinde değerlendirmemiz gerekiyor. Test kümesi, modelin yeni vakalarda ne kadar iyi performans gösterdiğini görmek için kullanabileceğimiz modelin eğitimden geçmediğine dair verdiğimiz verilerdir.

Eğitim performansının ölçülmesi durumunda, regresyon modellerinin test kümesi performansını değerlendirmek için bir karışıklık matrisi kullanamayız. Bunun yerine, modellerimizi değerlendirmek için bu bölümde daha sonra öğreneceğimiz Root Mean Squared Error (RMSE) gibi metrikleri kullanırız.

Modelimizin test kümesi performansını aldıktan sonra, tatmin edici olup olmadığına karar verebiliriz. Modelimizin harika bir iş çıkardığını düşünürsek, o zaman ortaya çıkardığı tahminlere güvenebiliriz ve bu tahminleri futbol takımımızın stratejisini yönetmek için kullanabiliriz. Ancak, modelimizin yeterince iyi olmadığını düşünürsek, 2. adıma (tahmin aşamasına) geri dönmeli ve diğer model alternatiflerini denemeliyiz.

# İçerik - Lineer Regresyon Varsayımları

Önceki derste, OLS'nin, modelin karesel hata koşullarının toplamını en aza indirerek doğrusal regresyon modelinin katsayılarını sıkıştırdığını gördük. Ancak, tahmin edilen katsayıların geçerli olması veya onlarla ilişkili test istatistiklerinin güvenilir olması için, veriler ve model ile ilgili bazı varsayımların karşılanması gerekir. Bu varsayımlar Gauss Markov Varsayımları veya Gauss Markov Koşulları olarak bilinir. Bu derste, bu varsayımları önceki dersteki tıbbi maliyetler modelimizi kullanarak inceleyeceğiz.

Doğrusal bir regresyon modelinin tahmin edilen katsayılarını yorumlamadan önce, Gauss Markov varsayımlarının geçerli olup olmadığını kontrol etmek her zaman iyi bir fikirdir. Aksi halde, modelimizi düzeltmeye çalışmalıyız. Bazen bu belirli bir problemi çözmek için bir teknik uygulamak anlamına gelir. Ancak genellikle, ek değişkenler ekleyerek veya problemli olanları hariç tutarak modelimizi değiştirmemiz gerekir. Modelimizi düzelttikten sonra, OLS kullanarak yeniden tahmin edebilir ve Gauss Markov koşullarının yerine getirilip getirilmediğini kontrol edebiliriz. Gördüğünüz gibi, bu yinelenen bir süreçtir.

Gauss Markov Koşullarının kesin sayısına dair bir not: Bazı yerlerde Gauss Markov Koşullarının sayısının dört, beş veya altı olduğunu görürseniz şaşırmayın! Bunun nedeni, bazı koşulların diğerlerinden türetilebilmesidir. Açıklık sağlamak için, Gauss Markov Koşullarını altı başlıkta tanıtmayı tercih ettik.

### Katsayılarda modelin doğrusallığı
Karşılanması gereken ilk varsayım, hedef değişkenin modelin lineer fonksiyonunun katsayısı olması gerektiğidir. İl bakışta bu durumdan, hedef ve özellikler arasındaki ilişkinin düz bir çizgi olma anlamı çıkarılabilir ancak bu durumdaherzaman geçerli değildir. İlişki, ikinci dereceden veya daha yüksek dereceli olabilir. Aşağıdaki gibi (denklem.1) gibi bir model tamamen geçerlidir:

### Hata terimi ortalamada sıfır olmalıdır
İkinci varsayım, hata teriminin beklenen değerinin sıfır olması gerektiğini belirtir. Matematiksel terimlerle ifade edecek olursak:

E(ϵ)=0
 
E  sembolü, beklenti operatörünü gösterir. Bunu Hata terimlerinin ortalaması sıfıra eşit olmalıdır şeklinde okuyabiliriz.

### Homoscedasticity
Üçüncü varsayım homoscedasticity gerekliliğidir. Bir model, hata koşullarının (scedasticity olarak bilinir) dağılımı, öngörülen tüm değerler için tutarlı olduğunda homoscedastictir. Başka bir deyişle, hata varyansı, gözlemler arasında sistematik olarak değişmemelidir. Bu varsayım karşılanmadığı zaman, heteroscesdastisity ile uğraşırız.

Örneğin, hata terimlerimiz sürekli olarak dağıtılmamışsa ve küçük sonuçlardan çok büyük sonuç değerleri için hatadan daha fazla varyansınız varsa, büyük tahmin edilen değerler için güven aralığı çok küçük olacaktır çünkü o ortalama hataya varyansı tabanlı olacaktır. Bu, modelimizin tahminlerinin doğruluğuna aşırı derecede güvenir.

### Düşük çoklu doğrusallık/low multicollinearity
Bireysel özellikler birbiriyle sadece zayıf bir şekilde ilişkili olmalı ve ideal olarak tamamen ilişkisiz olmalıdır. Özellikler ilişkilendirildiğinde, ikisi de sonuçtaki aynı varyans modelini açıklayabilir. Model, açıklayıcı gücün yarısını bir özelliğe ve diğerinin yarısına iliştirerek potansiyel bir çözüm bulmaya çalışacaktır. Tek amacımız tahmin olsaydı bu bir sorun olmayacaktı, çünkü önemli olan, varyansın açıklanmasıdır. Ancak, bir sonucu tahmin ederken en çok hangi özelliklerin önemli olduğunu bilmek istersek, çoklu-doğrusallık, özellikler ve sonuçlar arasındaki ilişkiyi küçümsememize neden olabilir.

Bir değişken ile bir başka veya birkaç değişken arasında 1 veya -1 arasında bir korelasyon varsa, buna mükemmel çoklu doğrusallık denir. İki değişken arasındaki mükemmel eşdoğrusallığı anlamak kolaydır. Ancak, bir değişken birkaç değişkenle nasıl ilişkilendirilebilir? Bu değişken diğerlerinin doğrusal bir kombinasyonu olduğunda gerçekleşir.

Dummy değişkenler hakkında bir not: Bu doğrusal kombinasyon sorunu nedeniyle dummy değişkenlerle çalışırken dikkatli olunmalıdır. Kategorik bir değişkenden bazı dummy değişkenler oluşturursak, bunlardan birini modelden hariç tutmamız gerekir. Çünkü bu dummy değişkenlerin herhangi biri, diğerlerinin toplamı eksi 1 olarak temsil edilebilir. Bu nedenle mükemmel bir çoklu doğrusallık meydana gelir.

Çoklu doğrusallığı saptamak için, özelliklerin korelasyon matrisine bakabiliriz. Çoklu doğrusallık, PCA tarafından veya ilişkili bazı özelliklerin atılmasıyla giderilebilir.

### Hata terimleri birbiriyle ilişkisiz olmalıdır
Hata terimleri birbiriyle ilişkisiz olmalıdır. Başka bir deyişle, bir gözlem için hata terimi, bir diğerinin hata süresini öngörmemelidir. Modelden ilgili bir değişkeni çıkarırsak, bu tür bir seri korelasyon olabilir. Dolayısıyla, bu değişkenin modele dahil edilmesi bu sorun için çözülebilir.

Hata koşullarının birbiriyle korele olup olmadığını belirlemek için bunları grafik olarak çizebiliriz. Grafikte rasgeleliğe dikkat etmeliyiz.

### Özellikler hatalarla korele olmamalıdır
Son fakat kesinlikle en önemlisi ve tartışmasız en önemli varsayım: açıklayıcı değişkenler ve hatalar bağımsız olmalıdır. Bu varsayım geçerli değilse, katsayıların tahminleri yanlı olacağından, modelin tahminleri güvenilmez olacaktır. Bu varsayım dışsallık/exogeneity olarak bilinir.

Dışsallık varsayımının ihlalleri çeşitli kaynaklara sahip olabilir. Yaygın nedenler ihmal edilen değişkenler ve bağımsız değişkenler ile hedef arasındaki eşzamanlı nedenselliktir. Sorun eşzamanlı nedensellikten kaynaklanıyorsa, sorunu çözmek için bazı ileri tekniklere başvurmalıyız, ancak bu, bu bootcampın kapsamı dışındadır.

### *Hataların normalliği hakkında çok önemli bir uyarı*
Şu ana kadar derste, OLS regresyonu için altı varsayımı ele aldık. Dikkate alınması gereken bir diğer önemli nokta, hata terimlerinin normalliğidir. Her ne kadar, bir OLS varsayımı olmasa da, yine de sonuçlarımızı etkileyebilir. Spesifik olarak, OLS'nin bir doğrusal regresyon modeline uygulanması için hataların normalliği gerekmemektedir, fakat tahmin edilen katsayılarımızın istatistiksel anlamlılığını ölçmek için, hata terimleri normal olarak dağıtılmalıdır. Bir sonraki derste hataların normalliğine dayanan t ve F testlerini ele alacağız.

Çoğu zaman, normal olarak dağıtılmamış hatalar ihmal edilen değişkenlerden kaynaklanır. Bu nedenle, modele ihmal edilen ilgili özellikleri eklemek sorunun düzeltilmesine yardımcı olabilir. Bazen bağımlı değişkeni dönüştürmek de yardımcı olur.

Hata koşullarının normalliğini kontrol etmek için çeşitli yollar vardır. Bunu yapmanın gayri resmi bir yolu, QQ çizimindeki hataları görselleştirmek veya histograma bakmaktır:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
weather_df = pd.read_csv("https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/weatherHistory.csv")
weather_df.head()
```

```python
y1 = weather_df["Temperature (C)"]
x1 = weather_df[["Humidity", "Wind Speed (km/h)", "Wind Bearing (degrees)", "Pressure (millibars)"]]
x1 = sm.add_constant(x1)
results1_model = sm.OLS(y1, x1)
results1 = results1_model.fit()
results1.summary()
```

```python
tahmin1 = results1.predict(x1)    # https://www.statsmodels.org/stable/examples/notebooks/generated/predict.html
hatalar1 = y1 - tahmin1

print(f"Hava durumu modelinin ortalama hatası: {hatalar1.mean():.16f}")
```

## Varsayım 3: homoscedasticity (eş varyanslık)

```python
plt.figure(figsize=(9,6), dpi=100)
plt.scatter(tahmin1, hatalar1)
plt.xlabel('Tahmin Edilen')
plt.ylabel('Artık (Residual)')
plt.axhline(y=0, color="r")
plt.title('Artık x Tahmin')
plt.show()
```

```python
from scipy.stats import bartlett
from scipy.stats import levene

bart_stats1 = bartlett(tahmin1, hatalar1)
lev_stats1 = levene(tahmin1, hatalar1)

print(f"Bartlett test değeri : {bart_stats1[0]:.3f} ve p değeri : {bart_stats1[1]:.21f}")
print(f"Levene test değeri   : {lev_stats1[0]:.3f} ve p değeri : {lev_stats1[1]:.21f}")
```

# İçerik - Hedef Değişken ve Özellikler Arasındaki İlişkiyi Anlama

Doğrusal regresyon modelinin gücü, sadece gelecekteki sonuçları öngörmekle kalmayıp, aynı zamanda bu sonuçları yaratan temel süreçleri de kavrayan, basit ve yorumlanabilir çözümler sağlama yeteneğinde yatmaktadır. Doğrusal regresyon modellerinin bu kadar yaygın olmasının nedenlerinden biri, basit bir şekilde, tahmin edilen katsayıları yorumlayarak, altında yatan ilişkiye ışık tutabilir. Tahmini katsayılar ve sonuçlar arasındaki ilişkiyi yorumlama sürecine çıkarım/interference denir.

Bu derste, aşağıdaki gibi soruları yanıtlamayı öğreneceğiz:

    Açıklayıcı bir değişkendeki artış, açıklanan değişkeni ne kadar etkiler?
    Belirli bir değişkenin etkisini, sonucu da etkileyen diğer değişkenlerin etkilerinden izole ederek nasıl seçebiliriz?
Bu soruların cevapları çok fazla iş değeri taşıyabilir. Reklam harcamalarının performansını ölçmeyi seven bir şirket düşünün. Radyo reklamları, yeni müşterileri çekmede Facebook reklamlarından daha etkili midir? TV reklamlarında harcanan bin dolar için şirket kaç tane ürün satıyor? Ayrıca, yeni müşterilerin TV reklamlarının veya Google reklamlarının sonucu olup olmadığını nasıl ayırt edebiliriz?

Bireysel özellikler ile hedef değişken arasındaki ilişkiyi ölçmeden önce, öncelikle bu özelliğin hedefin açıklanmasında "anlamlı/significant" bilginin olup olmadığını belirlememiz gerekir. Böylece istatistiksel anlamlılık tartışmasıyla başlayabiliriz.

Bu derste ele alacağımız ana konular şunlardır:

    İstatistiksel anlamlılık
    t-testi
    t-değeri
    p-değeri
    tahmin edilen katsayılar ile hedef arasındaki çıkarım ve ilişki
    sürekli ve kategorik değişkenlerin yorumlanması
    etkileşim terimlerini yorumlama
    sabitleri yorumlama
Bu dersin sonundaki ödevlerde de bazı modeller için katsayıları yorumlayacaksınız.

## Bireysel katsayıların istatistiksel anlamı
Doğrusal bir regresyon modelinde parametreleri tahmin etmek, tahmin edilen katsayıların istatistiksel olarak anlamlı olduğunu garanti etmez!

Bir parametreyi tahmin ederken nihai amaç, gerçek değerini popülasyonda bulmaktır. Daha sık olmamakla birlikte, gerçek popülasyondan ziyade bir örnekle çalışmamız gerekir ve bir örnek kullanarak katsayıları tahmin ederiz. Bir bireysel katsayının istatistiksel olarak anlamlı olabilmesi için, standart sapması açısından nispeten yüksek bir değere sahip olması ya da başka bir deyişle sonuçtaki bazı bilgileri açıklamaya yardımcı olması gerekir.

Bir bireysel katsayı tahmininin sıfırdan farklı olup olmadığını belirlemek için t-testi'ni kullanırız. Bu, parametrenin bir fonksiyonu olarak, diğer parametreler tarafından açıklanan varyansı kontrol ettikten sonra sonuçtaki benzersiz varyansın bir ölçüsünü almanın bir yoludur.

Örneğin, sarı dairenin sonucu gösterdiği doğrusal bir regresyonda, modelin açıkladığı toplam varyans, sarı, mavi ve pembe özellik dairelerinin kesişimi ile temsil edilir. Bununla birlikte, mavi dairenin öneminin istatistiksel testi sadece mavi, pembe ve sarının kesiştiği alanı göz ardı ederek, sadece mavi ve sarının kesiştiği yeşil alan üzerinde yapılacaktır. Benzer şekilde, pembe dairenin öneminin istatistiksel testi sadece pembe ve sarının kesiştiği turuncu alan üzerinde yapılacaktır.

Anlamlı olmayan/non-significant bir parametre, sonuç değişkeni hakkındaki bilgilerin açıklamasına katkıda bulunmaz ve modelin açıklayıcı gücünde çok az veya hiç azalma olmadan atılabilir. Ayrıca, istatistiksel olarak anlamsız katsayıları, modeli yorumlamada sıfır gibi ele almalıyız.

Statsmodels kullanarak bir doğrusal regresyon modelini tahmin ettiğimizde, summary() fonksiyonu t-istatistikleri ve ilişkili p-değerlerini döndürür. P-değerleri t-değerlerinden daha kolay okunur. Bir t-testi ile ilişkili p-değeri, gerçek popülasyonda tahmin edilen katsayıya aslında sıfıra eşit olma olasılığını ölçmektedir. P-değeri ne kadar düşük olursa, katsayı o kadar anlamlı olur. Genel bir kural olarak, bir katsayıya ait p-değeri 0.1'den küçük veya ona eşit olduğunda, katsayı istatistiksel olarak anlamlıdır deriz. Bununla birlikte, 0.05'e eşit veya daha küçük bir p-değeri her zaman daha iyidir.

## Kontrollü bir deney olarak OLS regresyonu
Kontrollü deney, sadece bir faktörün etkisinin, sonucu da etkileyen diğer faktörleri kontrol ederek ölçüldüğü bir bilimsel deney türüdür. Örneğin, ışığın belirli bir bitkinin büyümesi üzerindeki etkisini ölçmek istiyorsak, o zaman ışık miktarı haricinde nem, toprak gibi tüm faktörlerin aynı olduğu iki ortam kurmamız gerekir.

Tahmin edilen doğrusal regresyon modellerini kontrollü bir deney olarak düşünebiliriz. OLS kullanarak bir modelin katsayılarını tahmin ettiğimizde, aslında birbirlerinden izole edilen katsayı tahminleri elde ederiz. Yani, bir özellik katsayısı, tüm değişkenleri sabit tutarak, değişkenin sonuç değişkenine etkisini ölçer. Dolayısıyla, diğer tüm özelliklerin etkilerini yok ettikten sonra, tahmin edilen katsayıyı bu özellik ile hedef arasındaki ilişkinin ölçüsü olarak yorumlayabiliriz.

## Sürekli değişkenlerin yorumlanması
Şimdi, sürekli değişkenlerin katsayılarını (denklem.1) 'de yorumlayalım. İlk olarak, hem yaş hem de vki değişkenleri,  p -değerleri sıfıra çok yakın olduğundan istatistiksel olarak anlamlıdır. ## Etkileşim terimlerini yorumlama
Şimdiye kadar, kategorik ve sürekli değişkenleri birer birer yorumladık. Ancak bazen sonuçları en iyi şekilde açıklamak için etkileşimlerini modellemeliyiz. Doğrusal bir regresyon modelinde etkileşim değişkenleri koyarak, hedef ve özellikler arasındaki doğrusal olmayan ilişkileri yakalayabiliriz. Örneğin, (denklem.1)'den, sigara içenler ve içmeyenler arasındaki ortalama maliyet farkını biliyoruz. Ancak, sigara içenlerin maliyetinin, bir kişinin vücut kitle indeksi arttıkça (kalp krizi geçirme olasılığı artacağından) sigara içmeyenlerden daha fazla artacağına inanıyorsak, vki ile sigara_icer_mi değişkeni arasındaki etkileşimi modele eklememiz gerekir.

## Sabitin yorumlanması
Doğrusal regresyon modellerinde sabitler, tüm açıklayıcı değişkenler sıfır değerini aldığında gerçekleşmesi beklenen sonuç değerine karşılık gelir. Basit bir doğrusal regresyon ayarında (sadece tek bir açıklayıcı değişken olduğunda), sabitler, uygun eğrinin  y  eksenini kestiği noktaya karşılık gelir.

Sabit terimin basit doğrusal regresyon ayarlarında görselleştirilmesi kolay olsa da, modele dahil edilmiş birçok özellik varsa sabit bir terimin ne anlama geldiğini kavramak daha da güçleşir. Örneğin, bir modele bir dummy değişken dahil edilirse, daha önce açıkladığımız gibi bir kategori modelden hariç tutulmalıdır. Bu durumda, sabit terim de bekleme kategorisi ile ilgili bilgileri içerir. Bu nedenle, sabit tarafından çok sayıda faktör yakalanabilir.

Sabit terimi düşünmenin en güvenli yolu, bunu yanlı olarak düşünmektir. Bir önceki derste, sabit terimlerin, artanları ortalamada sıfır yaptığını hatırlayın. Bu anlamda, sabit yanlıyı artanlardan kendisine alır. Bu yüzden sabit de yanlı olarak adlandırılır!

## Bir modelin eğitim kümesindeki performansını ölçmek
R-kare değeri muhtemelen doğrusal regresyon modelinde uyumun iyi olması için en yaygın ölçüdür. Bu, modeldeki açıklayıcı değişkenler tarafından sonuç değişkeninde ne kadar varyans olduğunu açıklayan 0 ile 1 arasında bir orandır. Genel olarak, daha yüksek  R2  değerleri bir noktaya göre daha iyidir - düşük  R2 , modelimizin sonuç hakkında çok fazla bilgi vermediğini gösterir, bu da çok iyi tahminler vermeyeceği anlamına gelir. Ancak, çok yüksek  R2 , overfitting'in bir uyarı işaretidir. Hiçbir veri kümesi gerçekliğin mükemmel bir temsili değildir, bu nedenle verilerimize mükemmel uyan bir model ( R2  1 veya 1'e yakın), verilerdeki tuhaflıklar tarafından yanlıdır ve test kümesinde daha az performans gösterecektir. .

Yukarıdaki regresyon özeti tablosunda, tıbbi maliyetler modelimizin R-kare değerinin 0.747 olduğunu görüyoruz. Bu, modelimizin harcamalardaki varyansın %74,7'lik bir oranla açıkladığı, %25.3'ü ise açıklamadan bıraktığı anlamına gelir. İyileşme için hala bir yer olduğu sonucuna varabiliriz. Bir önceki dersteki Vücut kitle indeksi (vki) ve sigara_icer_mi etkileşimini içeren modeli tekrar uygulayalım:

```python
Y = sigorta['ucretler']
sigorta['vki_sigara'] = sigorta.vki * sigorta.sigara_icer_mi
X = sigorta[['kadin_mi','sigara_icer_mi', 'yas', 'vki', 'vki_sigara' ]]

X = sm.add_constant(X)

sonuclar = sm.OLS(Y, X).fit()
sonuclar.summary()
```

Bu modelin R-kare, önceki modelimizden daha yüksek olan 0.837'dir. Bu gelişme, vki ve sigara_icer_mi arasındaki etkileşimin, harcamalarda önceden açıklanmamış bir varyansı açıkladığını gösterir.

Daha önce söylediğimiz gibi, yüksek R-kare değerleri genellikle arzu edilir. Bununla birlikte, bazı durumlarda, çok yüksek R-kare değerleri, modelimizle ilgili bazı potansiyel sorunları göstermektedir. Özellikle:

Çok yüksek R-kare değeri, overfitting'in bir işareti olabilir. Eğer modelimiz veriler için çok karmaşıksa, o zaman eğitim kümesine overfit edebilir ve test kümesinde kötü bir iş çıkarabilir. Overfitting olduğunu tespit etmek için R-kare üzerinde anlaşılmış bir eşik yoktur. Bunun yerine, test ve eğitim verisindeki performans arasında bir karşılaştırma yapılmasını gerekir. Eğer modelimiz test kümesinde eğitim kümesine göre anlamlı derecede daha kötü bir performans sergiliyorsa, o zaman overfitting'den şüphelenmeliyiz. Bir sonraki derste test kümesindeki doğrusal regresyon modellerinin nasıl değerlendirileceğini tartışacağız.

R-kare performansın doğası gereği yanlı bir tahminidir. Modele daha açıklayıcı değişkenlerin eklenmesi, daha yüksek R-kare değerleri almamız anlamına gelir. Bu, sesler veya rastgele veriler gibi alakasız değişkenleri bile içerir. Bu sorunu hafifletmek için genellikle R-kare yerine ayarlanmış R-kare denilen bir ölçüm kullanırız. Ayarlanmış R-kare, R-kareyle aynı işi yapar, ancak modelde yer alan özelliklerin sayısına göre ayarlanır. Bu nedenle, R-kare değeri yerine ayarlanmış R-kare değerine bakmak her zaman daha güvenlidir.

Negatif R-kare değerleri üzerine bir not: Bazı modellerde negatif R-kare değerleri elde edilebilir. Genel anlamda, bir model düz bir yatay çizgiden daha zayıfsa, R-kare değeri negatif olur. Bu genellikle, modele bir sabit eklenmediğinde gerçekleşir. R-kare için negatif bir değer elde etmek, modelinizin hedefi açıklamada çok yetersiz kaldığı anlamına gelir.

# Farklı modellerin karşılaştırılması
Farklı modelleri karşılaştırmak ve en iyisini seçmek, veri bilimindeki en önemli uygulamalardan biridir. Çoğunlukla, birkaç modeli deneriz ve en iyi performansı belirlemek için bir test kümesindeki performanslarını değerlendiririz. Ancak, doğrusal regresyon modelleri söz konusu olduğunda çıkarım/interference da kritik bir görevdir. Tahmin gücünün test edilmesinden farklı olarak, çıkarımda, modellerimizin açıklayıcı gücünü önemseriz.

Bu derste, eğitim kümesinde modellerimizin performansını F-testi veya R-kare kullanarak ölçebileceğimizi gördük. Bu nedenle, F-testi ve R-kare, farklı modellerin karşılaştırılmasında kullanılabilir. Maalesef, bu iki ölçümün, belirli durumlarda kullanımı uygunsuz kılan bazı sakıncaları vardır.

Şimdi, model karşılaştırmasında F-testi ve R-kareyi nasıl kullanabileceğimize kısaca bakalım. Daha sonra, farklı modelleri karşılaştırmak için kullanabileceğimiz bilgi kriterlerine bakalım.

## 3.1. Model karşılaştırması için F-testinin kullanılması
Biri diğerinin içinde olan iki modeli karşılaştırmak için F-testi kullanabiliriz. Yani, bir modeldeki özellik kümesi diğerinin özellik kümesinin bir alt kümesi ise, F-testini kullanabiliriz. Bu durumda, daha yüksek F istatistiği olan modelin diğerine üstün olduğu söylenebilir.

Ancak, modeller içiçe değilse, F-testi kullanmak yanıltıcı olabilir. Hatalar normal dağılıma sahip değilse, diğer yöntemleri denemeliyiz.

## 3.2. Model karşılaştırması için R-kare kullanılması
Model karşılaştırması için R-kare de kullanılabilir. R-kare'nin, açıklayıcı değişkenlerin sayısı ile artma eğiliminde olduğu için yanlı olduğunu zaten gördük. Bu yüzden, R-kare yerine, ayarlanmış R-kare kullanabiliriz. Daha yüksek ayarlanmış R-kare, hedef değişkeni daha iyi açıklayan model anlamına gelir.

## 3.3. Bilgi kriterlerinin kullanılması
Bilgi kriterlerini kullanmak da farklı modelleri karşılaştırmak ve en iyisini seçmek için yaygın bir yoldur. Burada Akaike Information Criterion(AIC) ve Bayesian Information Criterion (BIC) olarak bilinen iki bilgi kriterinden bahsediyoruz. Her ikisi de kare hataları toplamını(SSE), örneklem boyutunu ve parametre sayısını dikkate alır.

AIC için formül:

nln(SSE)−nln(n)+2p
 
BIC için formül:

nln(SSE)−nln(n)+pln(n)
 
Bu formüllerin her ikisinde de,  n  örneklem boyutunu,  p  ise, modeldeki (sabit dahil) regresyon katsayılarının sayısını temsil eder.  ln  doğal logaritma anlamına gelir.

Hem AIC hem de BIC için, daha düşük değer daha iyidir. Bu nedenle, en düşük AIC veya BIC değeri olan modeli seçeriz. İki kriterden birini kullanabilmemize rağmen, AIC genellikle overfit eğilimi nedeniyle eleştirilmektedir. Aksine, BIC, parametrelerin sayısını AIC'den daha ciddi şekilde cezalandırır ve dolayısıyla daha cimri modellerin(yani, daha az parametreli modeller) lehine olur.

Ders boyunca, verilerimizi eğitim ve test kümelerine ayırarak tahminlerin nasıl yapılacağını göstereceğiz. Modellerimizi yalnızca eğitim kümesini kullanarak tahmin edeceğiz. Bunu yaparak, test kümesindeki modellerimizin değerlendirme sonuçlarını, gerçek dünya gözlemlerinde görünmeyen modellerimizin performansına göre okuyabiliriz.

## Doğrusal regresyon modellerini kullanarak tahmin yapmak
Tıbbi maliyet veri kümesinde nasıl tahmin yapacağımızı göreceğiz. Aynı şekilde verimiz hazırlayalım.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import statsmodels.api as sm
from statsmodels.tools.eval_measures import mse, rmse

%matplotlib inline
pd.options.display.float_format = '{:.3f}'.format

import warnings
warnings.filterwarnings(action="ignore")
```

```python
sigorta = pd.read_csv('https://djl-lms-assets.s3.eu-central-1.amazonaws.com/datasets/sigorta_verileri.csv')
sigorta.head()
```

```python
sigorta["kadin_mi"] = pd.get_dummies(sigorta.cinsiyet, drop_first=True)
sigorta["sigara_icer_mi"] = sigorta.sigara.replace({'evet':1, "hayır":0})
```

```python
Y = sigorta['ucretler']
X = sigorta[['kadin_mi','sigara_icer_mi', 'yas', 'vki' ]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 465)

print("Eğitim kümesindeki gözlem sayısı : {}".format(X_train.shape[0]))
print("Test kümesindeki gözlem sayısı   : {}".format(X_test.shape[0]))
```

```python
X_train = sm.add_constant(X_train)

sonuclar = sm.OLS(y_train, X_train).fit()

sonuclar.summary()
```

```python
X_test = sm.add_constant(X_test)

y_preds = sonuclar.predict(X_test)


baslik_font = {'family': 'arial','color':  'darkred','weight': 'bold','size': 15 }
eksen_font = {'family': 'arial','color':  'darkblue','weight': 'bold','size': 10 }
plt.figure(dpi = 100)

plt.scatter(y_test, y_preds)
plt.plot(y_test, y_test, color="red")
plt.xlabel("Gerçek Değerler", fontdict=eksen_font)
plt.ylabel("Tahmin edilen Değerler", fontdict=eksen_font)
plt.title("Ücretler: Gerçek ve tahmin edilen değerler", fontdict=baslik_font)
plt.show()

print("Ortalama Mutlak Hata (MAE)        : {}".format(mean_absolute_error(y_test, y_preds)))
print("Ortalama Kare Hata (MSE)          : {}".format(mse(y_test, y_preds)))
print("Kök Ortalama Kare Hata (RMSE)     : {}".format(rmse(y_test, y_preds)))
print("Ortalama Mutlak Yüzde Hata (MAPE) : {}".format(np.mean(np.abs((y_test - y_preds) / y_test)) * 100))
```

# İçerik - Aşırı Uyumluluk ve Regularizasyon

Bu derste, overfitting problemine ve bu problemi çözmek için genel yaklaşımlardan biri olan normalleştirme/regularization konusuna odaklanıyoruz. Bundan sonra, kayıp fonksiyonlarında normalizasyonu içeren doğrusal regresyon modellerinin üç varyantı olan Lasso, Ridge ve ElasticNet regresyonlarını inceleyecek ve aşağıdaki kavramları ele alacağız:

    overfitting
    underfitting
    genelleme/generalization
    normalleştirme/regularization
    Ridge regresyonu
    Lasso regresyonu
    ElasticNet regresyonu
Bu dersin sonunda, ev fiyatları modelinizi yeniden dönerek bir ödevi tamamlayacaksınız.

Overfitting
Makine öğrenimindeki temel zorluklardan biri, sadece bir eğitim kümesinde değil, aynı zamanda bir test kümesinde veya daha genel olarak görülmeyen gözlemlerde de performans gösteren modelleri tasarlamaktır. Daha somut olarak söylemek gerekirse:

modellerimizin eğitim kümesinde düşük hata oranı olmalı
Test kümesindeki ve eğitim kümesindeki hatalar arasındaki fark da düşük olmalı
Modelin bu ikinci koşulu yerine getirme yeteneği genelleme olarak adlandırılır ve test ve eğitim kümelerindeki hatalar arasındaki fark genelleme açığı olarak adlandırılır. İyi bir model iyi genellenmeli ve genelleme açığı mümkün olduğunca küçük olmalıdır.

Overfitting ve underfitting, makine öğrenmede temel kavramlardır ve bunları anlamak, iyi modellerin tasarlanması için çok önemlidir. Bir modelin genelleme açığı istediğimizden daha büyükse, buna overfitting diyoruz. Bir modelin eğitim performansı düşükse, biz buna underfitting diyoruz.

Bu bölümdeki doğrusal regresyon modellerinin uyumluluğunu tartışırken, aslında underfit olmayan modellere başvurmaya çalışıyorduk. Ancak, underfitting'ten kaçınmak iyi bir modelin seçiminin bir parçasıdır. İkinci parçası ise tahminle ilgilidir ve overfittingin tartışıldığı yer burasıdır.

Overfittingi daha iyi anlamak için, bir örnekle gösterelim. Aşağıda, iki değişken arasında ikinci dereceden bir ilişki oluşturacağız. Ardından, doğrusal regresyon modelleriyle ilişkisini özelliğin doğrusal, ikinci dereceden ve 10'uncu derece polinomunu kullanarak ayrı ayrı tahmin edeceğiz.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

import warnings
warnings.filterwarnings('ignore')
```

```python
np.random.seed(456)
X = pd.DataFrame(25 - np.arange(1,51)) 
```

```python
Y = X + X**2 +  pd.DataFrame(np.random.normal(0,100,50))
```

```python
lrm_underfit = LinearRegression().fit(X, Y)
underfit_tahmin = lrm_underfit.predict(X)

# Kare Fonksiyon için model
X_dogru = pd.concat([X, X**2], axis=1)
lrm_correct = LinearRegression().fit(X_dogru, Y)
dogru_tahmin = lrm_correct.predict(X_dogru)

# 10'uncu derece fonksiyon için model
X_overfit = pd.concat([X**i for i in range(1,11)], axis=1)
lrm_overfit = LinearRegression().fit(X_overfit, Y)
overfit_tahmin = lrm_overfit.predict(X_overfit)
```

```python
plt.figure(figsize=(30,10))
baslik_font = {'color':  'darkred','weight': 'bold','size': 30 }
eksen_font = {'color':  'darkblue','weight': 'bold','size': 20 }

plt.subplot(1,3,1)
plt.scatter(X, Y, color="red")
plt.plot(X.values, underfit_tahmin)
plt.title("Underfitting", fontdict=baslik_font)
plt.xlabel("Özellik (X)", fontdict=eksen_font)
plt.ylabel("Hedef (Y)", fontdict=eksen_font)

plt.subplot(1,3,2)
plt.scatter(X.values, Y, color="red")
plt.plot(X.values, dogru_tahmin)
plt.title("Doğru ilişki", fontdict=baslik_font)
plt.xlabel("Özellik (X)", fontdict=eksen_font)
plt.ylabel("Hedef (Y)", fontdict=eksen_font)

plt.subplot(1,3,3)
plt.scatter(X.values, Y, color="red")
plt.plot(X.values, overfit_tahmin)
plt.title("Overfitting", fontdict=baslik_font)
plt.xlabel("Özellik (X)", fontdict=eksen_font)
plt.ylabel("Hedef (Y)", fontdict=eksen_font)

plt.show()
```

İlk grafik underfittingi gösterir. Doğru ilişki ikinci dereceden olduğundan, düz bir çizgi ayarlayarak, model hedefte varyasyonu açıklayan kötü bir iş yapar.

Üçüncü grafik overfittingi göstermektedir. 10'uncu derece polinom verideki gürültüye uyuyor. Üçüncü grafiğe yerleştirilmiş mavi çizgi 'nin en sağ kısmına bakarsak, hattın dönmeye başladığını görürüz. Bununla birlikte, özelliğin büyük değerleri için hedef değerlerin tümü grafiğin sağ üst köşesine doğru gelmeye devam edecektir. Bu nedenle, üçüncü grafikteki model bu değerleri doğru bir şekilde tahmin etmeyecektir.

İkinci grafik, veriler arasındaki gerçek ilişkiyi göstermektedir. Uygun olan çizgi gürültüye karşı dayanıklıdır ve doğru ilişkiyi yakalar.

Genel bir kural olarak, eğer modelimiz gereğinden daha karmaşıksa, overfit eğilimi gösterecektir. Tersine, eğer modelimiz gereğinden az karmaşıksa, o zaman eğitim kümesine underfit olacaktır.

Bir modelin karmaşıklığının aynı zamanda kapasitesi olarak adlandırıldığını unutmayın. Dolayısıyla, modelimizin overfit olduğunu tespit edersek, modelin karmaşıklığını veya kapasitesini azaltmalıyız. Bunu, modelden önemsiz olan bazı özellikleri ortadan kaldırarak veya polinomun derecesini azaltarak yapabiliriz. Ayrıca, overfitting ile başa çıkmak için bazı genel yaklaşımlar da vardır. Bunlardan biri, bu derste ele alacağımız normalleşmedir.

Aşağıda, ev fiyatları veri kümesini kullanacağız ve verileri overfit hale getirmek için çok karmaşık bir model oluşturacağız. Özellikle, verilerdeki tüm sayısal özellikleri ve üslerini 21'e kadar kullanacağız.

## Regularization
Modellerimizi test kümesi üzerinde daha iyi genellemek için, bazen modellerin öğrenme hedeflerini veya kayıp fonksiyonlarını ayarlamak gerekmektedir. Bunu yaparak, tercihlerimizi potansiyel çözümler üzerinde empoze ediyoruz ve tercih edilmeyen çözümlerden birini daha iyi performans göstermediği sürece modelleri tercih edilen çözümlerimizden birini seçmeye zorluyoruz. Genel olarak, normalleştirme/regularization terimi, eğitim performansından ödün vermeden genelleme açığını azaltmak için algoritmaların modifiye edilmesini ifade eder.

Doğrusal regresyon tanımladığımızda, model uyumunun, öngörülen ve gerçek değerler arasındaki karesel farkların toplamını en aza indirerek belirlendiğini ve buna Ordinary Least Squares dendiğini görmüştük:


 
Ancak, bu maliyet fonksiyonunu çeşitli şekillerde değiştirerek daha doğru tahminler elde edebiliriz. Bunu yapmanın bir yolu, OLS maliyet fonksiyonunun eğitim kümesindeki açıklanmış varyansı optimize etmesidir. Ridge, Lasso ve ElasticNet regresyonları bu maliyet fonksiyonunu değiştirmemize olanak sağlar. Bunlar test kümelerinde açıklanan varyansı optimize eden modeller ortaya çıkmasına yol açarlar. Genel olarak amacımız, sadece eğitim kümesi için değil karşılaşılabilecek tüm veriler hakkında bize bilgi veren bir model çıkarmaktır, bu yüzden Ridge, Lasso ve ElasticNet çözümleri bize oldukça yardımcı olmaktadır.

## Ridge regresyonu
Ridge regresyonu maliyet fonksiyonunu en aza indirir:

Bu yeni bölüm, büyük katsayılar için bir ceza uygular. Tüm model katsayılarının(1'den  p 'a kadar) karelerinin toplamı,  λ  değişkenlik parametresiyle çarpılır.  λ 'nın büyüdükçe, katsayı büyüklüğünün cezası da artar ve maliyet fonksiyonunu en aza indiren çözümün daha küçük katsayı tahminleri olacaktır. Kare ağırlıkların toplamına göre düzenli hale getirme de "L2 normalizasyon" olarak adlandırılır.

Lambda adlandırması ile ilgili bir not: Eğer konvansiyonelleştirme parametresi "lambda" ( λ ) olarak adlandırılacaksa, scikit-learn neden buna "alpha" diyor? Çünkü lambda Python'da ayrılmış bir anahtar kelimedir!

Ridge regresyonunun arkasındaki temel ilke, modeller gittikçe karmaşıklaştığında ve özellikler birbirleriyle daha fazla ilişkili olduğunda (multicollinear), OLS ile gelen katsayıların giderek daha büyük hale gelmesidir. Bu, modelin eğitim kümesinden çok fazla varyansı içerdiğini gösteren bir işarettir - başka bir deyişle, overfittingdir. Ridge regresyonu bir model varyansını minimizeetmek için çalışır.

Şimdi, önceki modelimizin overfittingiyle başa çıkmak için konut fiyatları veri kümesinde bir Ridge regresyon modeli oluşturuyoruz. Yukarıdaki doğrusal regresyon modeliyle aynı özellikleri kullanıyoruz:

```python
from sklearn.linear_model import Ridge

ridgeregr = Ridge(alpha=10**37) 
ridgeregr.fit(X_egitim, y_egitim)

y_egitim_tahmini = ridgeregr.predict(X_egitim)
y_test_tahmini = ridgeregr.predict(X_test)

print("Eğitim kümesi R-Kare değeri       : {}".format(ridgeregr.score(X_egitim, y_egitim)))
print("-----Test kümesi istatistikleri---")
print("Test kümesi R-Kare değeri         : {}".format(ridgeregr.score(X_test, y_test)))
print("Ortalama Mutlak Hata (MAE)        : {}".format(mean_absolute_error(y_test, y_test_tahmini)))
print("Ortalama Kare Hata (MSE)          : {}".format(mse(y_test, y_test_tahmini)))
print("Kök Ortalama Kare Hata (RMSE)     : {}".format(rmse(y_test, y_test_tahmini)))
print("Ortalama Mutlak Yüzde Hata (MAPE) : {}".format(np.mean(np.abs((y_test - y_test_tahmini) / y_test)) * 100))
```

Eğitim kümesinin R-kare değeri 0,95'ten 0,67'ye düştü. Bu bir bozulma gibi görünse de, test kümesinde R-kare değeri bakarsak, şimdi 0.30 olduğunu görürüz ki bu da negatif bir sıçrama anlamına gelir. Dahası, test kümesindeki tüm performans istatistikleri oldukça önemli ölçüde iyileşti. Bunlar, Ridge regresyonunu kullanarak, overfittingin büyüklüğünü azalttık anlamına gelir.

Ridge'in regülasyon parametresinin ayarlanması
Yukarıdaki modelde,  λ=1037  normalleşme parametresi ayarladık. Ridge regresyonunda,  λ  olan normalleşme parametresi, 0'dan büyük herhangi bir değeri alabilir. Optimal bir normalleşme parametresini seçmenin en iyi yolu, çapraz doğrulamayladır. Bu hangi parametrenin eğitim ve doğrulama kümelerinde en tutarlı sonuçları verdiğini kontrol eder. validasyon kümesi, çapraz doğrulama uygularken eğitim kümesinden ayrı tuttuğumuz kümedir.

## Lasso regresyonu
Ridge regresyonu, parametre tahminlerini küçültür, ancak tahminler asla tam olarak 0'a ulaşmaz. Öte yandan LASSO(En Az Mutlak Küçüllme ve Seçme Operatörü/Least Absolute Shrinkage and Selection Operator), küçük parametre tahminlerini onları modelden etkin bir şekilde düşürerek sıfırlamak için zorlamaya çalışan bir model optimizasyon tekniğidir. Bu, overfittingi önleyebilir ve gömülü özellik seçme yöntemi olarak da kullanılabilir. Lasso, binlerce veya yüz binlerce tahminciyle uğraşırken ve işlemci süresini optimize etmeniz gerektiğinde veya daha kolay yorumlanabilecek daha basit bir çözüme ulaşmak istediğinizde son derece kullanışlıdır.

Lasso için en aza indirgeyecek maliyet fonksiyonu, Ridge için en aza indirilen maliyet fonksiyonuna çok benzerdir:


```python
from sklearn.linear_model import Lasso

lassoregr = Lasso(alpha=10**20.5) 
lassoregr.fit(X_egitim, y_egitim)

y_egitim_tahmini = lassoregr.predict(X_egitim)
y_test_tahmini = lassoregr.predict(X_test)

print("Eğitim kümesi R-Kare değeri       : {}".format(lassoregr.score(X_egitim, y_egitim)))
print("-----Test kümesi istatistikleri---")
print("Test kümesi R-Kare değeri         : {}".format(lassoregr.score(X_test, y_test)))
print("Ortalama Mutlak Hata (MAE)        : {}".format(mean_absolute_error(y_test, y_test_tahmini)))
print("Ortalama Kare Hata (MSE)          : {}".format(mse(y_test, y_test_tahmini)))
print("Kök Ortalama Kare Hata (RMSE)     : {}".format(rmse(y_test, y_test_tahmini)))
print("Ortalama Mutlak Yüzde Hata (MAPE) : {}".format(np.mean(np.abs((y_test - y_test_tahmini) / y_test)) * 100))
```

R-kare, eğitim kümesinde 0.75 ve test kümesinde 0,51'tir. Test kümesindeki R kare değeri, bu derste ele aldığımız modeller arasında en yüksek olanıdır. Ayrıca, eğitimin ve test kümesinin R-kare değeri arasındaki fark da en düşüktür. Test kümesindeki performans istatistiklerinin tümü de en düşüktür ve bu Lasso modelinin daha önce yaptığımız Ridge regresyon ve lineer regresyon modellerinden daha iyi performans gösterdiği anlamına gelir.


Lasso'nun normalleştirme parametresini ayarlama
Lasso için  λ  0 (sıfır ceza, OLS gibi davranır) ve sonsuzluk arasında değişebilir.  λ  değeri çok büyükse, tüm parametreler sıfırlanır. Her zamanki gibi  λ  parametresini belirlemenin en iyi yolu çapraz doğrulamadır!

## ElasticNet regresyonu
Bu derste daha önceki tartışmalarımızda da vurguladığımız gibi, hem Lasso hem de Ridge regresyonlarının kendi artıları ve eksileri vardır. Lasso'nun L1 regülasyonunu ve Ridge'in L2 regülasyonunu birleştiren popüler bir yönteme ElasticNet regresyon denir. ElasticNet regresyonunun kayıp fonksiyonu:
λ1  sıfıra eşit olduğunda, bu kayıp fonksiyonunun Ridge regresyonunun kayıp fonksiyonuna dönüştüğünü ve  λ2  sıfıra eşit olduğunda Lasso regresyonunun kayıp fonksiyonu haline geldiğine dikkat edin. Bu nedenle, hem Lasso hem de Ridge regresyonları, ElasticNet regresyonunun özel bir durumudur. Dahası, hem  λ1  ve  λ2  sıfıra eşit olduğunda, doğrusal regresyon modeline dönüşür!

Şimdi, ev fiyatları modelinde ElasticNet regresyonunu deneyelim:

```python
from sklearn.linear_model import ElasticNet

elasticregr = ElasticNet(alpha=10**21, l1_ratio=0.5) 
elasticregr.fit(X_egitim, y_egitim)

y_egitim_tahmini = elasticregr.predict(X_egitim)
y_test_tahmini = elasticregr.predict(X_test)

print("Eğitim kümesi R-Kare değeri       : {}".format(elasticregr.score(X_egitim, y_egitim)))
print("-----Test kümesi istatistikleri---")
print("Test kümesi R-Kare değeri         : {}".format(elasticregr.score(X_test, y_test)))
print("Ortalama Mutlak Hata (MAE)        : {}".format(mean_absolute_error(y_test, y_test_tahmini)))
print("Ortalama Kare Hata (MSE)          : {}".format(mse(y_test, y_test_tahmini)))
print("Kök Ortalama Kare Hata (RMSE)     : {}".format(rmse(y_test, y_test_tahmini)))
print("Ortalama Mutlak Yüzde Hata (MAPE) : {}".format(np.mean(np.abs((y_test - y_test_tahmini) / y_test)) * 100))
```

## Model 1: OLS Regresyonu
Bir önceki derste en iyi açıklama yüzdesine sahip model kullanılacaktır.
Değişkenler: ["TotalBsmtSF", "OverallQual", "GrLivArea", "YearBuilt"]

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge, Lasso, ElasticNet, RidgeCV, LassoCV, ElasticNetCV
    # https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html
    # https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Lasso.html
    # https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.ElasticNet.html
    # https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.RidgeCV.html
    # https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LassoCV.html
    # https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.ElasticNetCV.html
from sklearn.model_selection import KFold, cross_validate, cross_val_predict
    # https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html
    # https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_validate.html
    # https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_predict.html
import statsmodels.api as sm
from statsmodels.tools.eval_measures import mse, rmse

%matplotlib inline
sns.set(style = "whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

title_font = {"family":"arial", "weight":"bold", "color":"darkred", "size":16}
axis_font = {"family":"sans", "weight":"normal", "color":"darkblue", "size":14}

house = pd.read_csv("house_prices.csv")
```

```python
# Modelimizi oluşturalım:
x1 = house[["TotalBsmtSF", "OverallQual", "GrLivArea", "YearBuilt"]]
y1 = house["SalePrice"]
```

```python
kFold = KFold(10, shuffle=True, random_state=42)
mae1, mse1, rmse1, mape1, trainR_mean, testR_mean = ([] for i in range(6))

for train, test in kFold.split(x1, y1): 
    x1_train = x1.iloc[train]
    x1_test = x1.iloc[test]
    y1_train = y1.iloc[train]
    y1_test = y1.iloc[test]
    
    # Eğitim verilerinin R-kare değeri
    x1_train = sm.add_constant(x1_train)
    results1_model = sm.OLS(y1_train, x1_train)
    results1_train = results1_model.fit()
    trainR_mean.append(results1_train.rsquared_adj)

    # Test verilerinin R-kare değeri
    x1_test = sm.add_constant(x1_test)
    results1_model = sm.OLS(y1_test, x1_test)
    results1_test = results1_model.fit()
    testR_mean.append(results1_test.rsquared_adj)

    y1_predict = results1_train.predict(x1_test)

    mae0 = mean_absolute_error(y1_test, y1_predict)
    mse0 = mse(y1_test, y1_predict)
    rmse0 = rmse(y1_test, y1_predict)
    mape0 = (abs((y1_test - y1_predict)) / y1_test).mean() * 100
    
    mae1.append(mae0)
    mse1.append(mse0)
    rmse1.append(rmse0)
    mape1.append(mape0)

print(f"Eğitim verilerinin ayarlanmış R-kare değeri : {np.array(trainR_mean).mean():.3f}")
print(f"Test verilerinin ayarlanmış R-kare değeri   : {np.array(testR_mean).mean():.3f}")
print("\n---------Test Kümesi Hata İstatistikleri---------")
print(f"Ortalama Mutlak Hata (MAE)        : {np.array(mae1).mean():.2f}")
print(f"Ortalama Kare Hata (MSE)          : {np.array(mse1).mean():.2f}")
print(f"Kök Ortalama Kare Hata (RMSE)     : {np.array(rmse1).mean():.2f}")
print(f"Ortalama Mutlak Yüzde Hata (MAPE) : %{np.array(mape1).mean():.2f}")
```
